````python
import os, requests

url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-flash-lite-latest:generateContent"
headers = {"Content-Type": "application/json", "X-goog-api-key": os.getenv("GEMINI_API_KEY")}
data = {"contents": [{"parts": [{"text": "Write a short poem on ME"}]}]}

response = requests.post(url, headers=headers, json=data)
print(response.json()['candidates'][0]['content']['parts'][0]['text'])
````

# Project: Calling Gemini API via Raw HTTP

I wrote a lightweight Python script to send direct network requests to Google's Gemini models without relying on their official developer SDK.

## What this code does:
1. **Imports Core Modules:** Uses `os` to safely load environment credentials and `requests` to handle HTTP protocols.
2. **Configures the Endpoint:** Targets the production `gemini-flash-lite-latest` text generation API route.
3. **Passes Authentication:** Reads the `GEMINI_API_KEY` token dynamically from the system environment variables and passes it securely through the `X-Goog-Api-Key` header.
4. **Submits Payload:** Formats a structured JSON dictionary containing the instruction prompt (`"Write a short poem on ME"`).
5. **Parses Output:** Extracts and prints the direct string output by drilling into the deeply nested JSON array structure: `['candidates'][0]['content']['parts'][0]['text']`.

---

````python
import os, google.genai as genai

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
print(client.models.generate_content(model='gemini-flash-lite-latest', contents='Hi').text)
````
